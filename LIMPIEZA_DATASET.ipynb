# Primera parte del preprocesamiento. Limpieza del dataset previo al split.

```python
import pandas as pd
from preprocessing.data_cleanse import limpiar_dataset

dev = pd.read_csv('data/raw/pf_suvs_i302_1s2025.csv')

dev = limpiar_dataset(dev)
dev.to_csv('data/processed/pf_suvs_i302_1s2025_cleaned.csv', index=False)
```

```python
cleaned = pd.read_csv('data/processed/pf_suvs_i302_1s2025_cleaned.csv')

#datos diferentes en cada columna
print("Datos diferentes en cada columna:")
print(cleaned.nunique())

print("\nDatos nulos en cada columna:")
print(cleaned.isnull().sum())
```

# Encaro la segunda parte del preprocesamiento. Ya con el dataset limpio y prolijo, manejo los valores faltantes. Se aprende desde el train, y se aplica a val y test. 


```python
#spliteo cleaned en train, test y validation
from sklearn.model_selection import train_test_split
#train debe ser el 64%,val el 16% y test el 20%
train, test = train_test_split(cleaned, test_size=0.2, random_state=42)
val, test = train_test_split(test, test_size=0.25, random_state=42)
#guardar train, test y val
train.to_csv('data/processed/df_train.csv', index=False)
test.to_csv('data/processed/df_test.csv', index=False)
val.to_csv('data/processed/df_val.csv', index=False)
```
